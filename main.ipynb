```python

from google import genai
from google.genai.types import ReplayResponse
import requests
from PIL import Image
from io import BytesIO
import json
import time


# Replace with your actual Gemini API key
API_KEY = "[REMOVED]"

# Initialize the new SDK client for text generation.
client = genai.Client(api_key=API_KEY)
```

```python
# Cell 3: Define Text and Image Generation Functions
# -----------------------------------------------
def generate_text_content(ingredients: str) -> str:
    """
    Generates marketing content using Gemini-2.0.
    """
    prompt = (
        f"You are an AI agent for Marketing. "
        f"Develop a platform for generating marketing slogans, ad copy, and campaign ideas. "
        f"Input details: {ingredients}"
    )
    # Using Gemini-2.0 flash model for text
    resp: ReplayResponse = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=prompt
    )
    return resp.text


# Function to generate images using Craiyon (DALL·E Mini)
def generate_images(prompt: str):
    """
    Generates images using Craiyon (DALL·E Mini) with browser-like headers.
    """
    try:
        API_URL = "https://backend.craiyon.com/generate"
        payload = {"prompt": prompt}
        headers = {
            "Content-Type": "application/json",
            "User-Agent": ("Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                           "AppleWebKit/537.36 (KHTML, like Gecko) "
                           "Chrome/115.0.0.0 Safari/537.36"),
            "Referer": "https://craiyon.com/"  # Optional: set the referer to the Craiyon site
        }
        
        print("Generating image... This may take up to 30 seconds.")
        response = requests.post(API_URL, headers=headers, data=json.dumps(payload))
        
        if response.status_code == 200:
            data = response.json()
            images = data.get("images", [])
            if images:
                img_list = []
                for idx, img_data in enumerate(images):
                    # Craiyon returns images as hexadecimal strings.
                    img_bytes = bytes.fromhex(img_data)
                    img = Image.open(BytesIO(img_bytes))
                    # Save image locally and display it
                    img_path = f"generated_image_{idx}.png"
                    img.save(img_path)
                    img_list.append(img)
                    display(img)
                return img_list
            else:
                print("No images returned from Craiyon.")
                return []
        else:
            print(f"Error {response.status_code}: {response.text}")
            return []
    except Exception as e:
        print(f"Error generating images: {e}")
        return []
```

```python
# Cell 4: Test Text Generation
# -----------------------------------------------
ingredients = "marketing slogans, ad copy, campaign ideas"
generated_text = generate_text_content(ingredients)
print("Generated Marketing Content:\n")
print(generated_text)
```
